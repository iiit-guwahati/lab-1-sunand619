```python
summ=0
import math
print(len(arr))
for i in range(len(arr)):
  summ+=arr[i]
print('Sum: '+str(summ)) 
mean=summ/len(arr) 
print('Mean: '+str(summ/len(arr)))
arr.sort()
size=len(arr)
if(size%2==0):
  print('Median: '+str((arr[size//2-1]+arr[size//2])//2))
else:
  print('Median: '+str(arr[(size+1)//2-1]))  
sum=0
for i in range(len(arr)):
  sum+=(arr[i]-mean)**2
print('Std: '+str(math.sqrt(sum/len(arr))))   
```

# Working with CSV

Let's import a file with fuel economy data

```python
import csv
with open('mpg.csv') as csvfile:
    mpg = list(csv.DictReader(csvfile))
    
mpg[:3] 
```

Find out how many items are there in the dictionary

```python
print(len(mpg[0]))
```

Find the column names with keys

```python
for i in mpg[0]:
  print(i)
```

Find the average cty fuel economy across all cars. All values in the dictionaries are strings.

```python
sum=0
for i in range(len(mpg)):
  dict=mpg[i]
  sum+=int(dict['cty'])
print(sum/len(mpg))
```

Find the average hwy fuel economy across all cars

```python
sum=0
for i in range(len(mpg)):
  dict=mpg[i]
  sum+=int(dict['hwy'])
print(sum/len(mpg))
```

Use set to return the unique values for the number of cylinders the cars.

```python
cylinders=set()
for i in range(len(mpg)):
  dict=mpg[i]
  cyl=int(dict['cyl'])
  cylinders.add(cyl)
print(cylinders)    
```

Group the cars by number of cylinder, and finding the average cty mpg for each group.

```python
cylinders=list(cylinders)
MpgByCyl = []
for i in cylinders: # iterate over all the cylinder levels
    sumpg = 0
    typecount = 0
    for d in mpg: # iterate over all dictionaries
      if i==int(d['cyl']):
        sumpg+=int(d['cty'])
        typecount+=1
    MpgByCyl.append(sumpg/typecount) 
print(MpgByCyl)       
        
```

Return the unique values for the class types in our dataset using set.

```python
vehicleclass = set(d['class'] for d in mpg)
print(vehicleclass)
```

Find average miles per gallon in the highway for each class of vehicles

```python
MileageByClass = []
for t in vehicleclass: # iterate over all the vehicle classes
    sumpg = 0
    classCount = 0
    for d in mpg: # iterate over all dictionaries
       if d['class']==t:
         sumpg+=int(d['hwy'])
         classCount+=1
    MileageByClass.append(sumpg/classCount)
print(MileageByClass)         
```
